```python
def neg_log_likelihood(sigma2, U, V, users, jokes, ratings):
    constant = Variable((torch.ones(1) * np.log((2*np.pi*sigma2)**0.5)).double())
    multiplier = Variable((torch.ones(1) * (2*sigma2)**(-1.0)).double())
    return constant + multiplier * torch.sum((ratings.double() - torch.mm(U, torch.transpose(V, 0, 1))[users.data, jokes.data])**2)

def U_grad(sigma2, K, U, V, users, jokes, ratings_matrix):
    multiplier = Variable((-torch.ones(1) * (sigma2)**(-1.0)).double())
    U_grad_matrix = np.zeros((users.size()[0], U.size()[1]))
    for i in range(users.size()[0]):
        U_grad_matrix[i] = np.sum(np.multiply(V[jokes].data.numpy(), (ratings.double() - torch.mm(U, torch.transpose(V, 0, 1))[users.data, jokes.data]).data.numpy().reshape(1000, -1)), axis = 0)
    return U_grad_matrix

def V_grad(sigma2, U, V, users, jokes, ratings_matrix):
    multiplier = Variable((-torch.ones(1) * (sigma2)**(-1.0)).double())
    V_grad_matrix = np.zeros((jokes.size()[0], V.size()[1]))
    for i in range(jokes.size()[0]):
        V_grad_matrix[i] = np.sum(np.multiply(V[jokes].data.numpy(), (ratings.double() - torch.mm(U, torch.transpose(V, 0, 1))[users.data, jokes.data]).data.numpy().reshape(1000, -1)), axis = 0)
    return U_grad_matrix
```

```python
train_iter, val_iter, test_iter, text_field = load_jester(ratings_path='jester_ratings.dat.gz', jokes_path='jester_items.clean.dat.gz', batch_size=1000, gpu=False, repeat=False, shuffle=True)
```

```python
K = 2
sigma2 = 1.0
U = Variable(torch.from_numpy(np.random.rand(ratings_matrix.shape[0], K).reshape(ratings_matrix.shape[0], K)), requires_grad = True)
V = Variable(torch.from_numpy(np.random.rand(ratings_matrix.shape[1], K).reshape(ratings_matrix.shape[1], K)), requires_grad = True)
optimizer = torch.optim.SGD([U, V], lr = 0.05)

batch_num = 1
for batch in train_iter:
    print(batch_num)
    batch_num += 1
    ratings = batch.ratings-1 
    users = batch.users-1 
    jokes = batch.jokes-1
    break
    
    optimizer.zero_grad()
    loss = neg_log_likelihood(sigma2, U, V, users, jokes, ratings)
    loss.backward()
    optimizer.step()
    
    
    
```
